```python
import requests
from bs4 import BeautifulSoup
import csv

quotes = []
url = "https://quotes.toscrape.com"

while url:
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")

    for q in soup.find_all("div", class_="quote"):
        text = q.find("span", class_="text").text
        author = q.find("small", class_="author").text
        if author == "Albert Einstein":
            quotes.append(text)

    next_btn = soup.find("li", class_="next")
    url = "https://quotes.toscrape.com" + next_btn.find("a")["href"] if next_btn else None

with open("quotes_Einstein.csv", "w", encoding="utf-8-sig", newline="") as f:
    writer = csv.writer(f)
    for q in quotes:
        writer.writerow([q])

print("Saved quotes_Einstein.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import csv

life_quotes = []
url = "https://quotes.toscrape.com"

while url:
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")

    for q in soup.find_all("div", class_="quote"):
        text = q.find("span", class_="text").text
        if "life" in text.lower():
            life_quotes.append(text)

    next_btn = soup.find("li", class_="next")
    url = "https://quotes.toscrape.com" + next_btn.find("a")["href"] if next_btn else None

with open("life_quotes.csv", "w", encoding="utf-8-sig", newline="") as f:
    writer = csv.writer(f)
    for q in life_quotes:
        writer.writerow([q])

print("Saved life_quotes.csv")
```
